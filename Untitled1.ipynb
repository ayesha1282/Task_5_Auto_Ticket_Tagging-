```python
tickets = [
    "I cannot log into my account",
    "I want a refund for my subscription",
    "The app crashes when I open it",
    "My payment was charged twice",
    "I forgot my password"
]

tickets
```

```python
possible_tags = [
    "Login Issue",
    "Billing Issue",
    "Technical Problem",
    "Account Management",
    "Payment Issue"
]

possible_tags
```

```python
def zero_shot_tag(ticket):
    prompt = f"""
You are a support assistant.
Ticket: "{ticket}"
Choose the best 3 tags from this list:
{possible_tags}
Return only the tags.
"""
    return prompt
```

```python
print(zero_shot_tag(tickets[0]))
```

```python
few_shot_prompt = """
Examples:
Ticket: "I forgot my password"
Tags: Login Issue, Account Management

Ticket: "My card was charged twice"
Tags: Billing Issue, Payment Issue

Now tag this ticket:
"""
```

```python
def few_shot_tag(ticket):
    prompt = few_shot_prompt + f'\nTicket: "{ticket}"\nTags:'
    return prompt
```

```python
for t in tickets:
    print("Ticket:", t)
    print("Zero-shot prompt:", zero_shot_tag(t))
    print("Few-shot prompt:", few_shot_tag(t))
    print("-" * 40)
```

```python
"""
Conclusion:
- Zero-shot tagging works but may be inaccurate
- Few-shot learning improves tag relevance
- LLMs can classify tickets without traditional ML training
"""
```
